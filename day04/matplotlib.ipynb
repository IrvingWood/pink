### <b>设置子画布</b>

```python
# 画板——>画布——>呈现图像
```

```python
# 画布和画板一样，如果不需要做一些额外设置，不需要显示调用它
# 常用操作是对画板进行分区，就要配置画布的位置
axes=plt.subplot()
```

```python
# 调整画布的大小，可以通过调整画板来实现
plt.figure(figsize=(10,6))
axes1=plt.subplot(221)
# 在不同的画布中，绘制不同的图像
# 1，上下文原则：绘制函数会默认把图像绘制到离他最近的那个画布中
plt.plot(x,np.sin(x))
# 2,使用面向对象的方式绘制
axes1.plot(x,np.cos(x))

axes2=plt.subplot(222)

axes3=plt.subplot(223)

axes4=plt.subplot(224)
plt.plot(x,np.sin(x))
```

```python
plt.grid()
```

```python
plt.figure(figsize=(10,6))
axes1=plt.subplot(221)
plt.grid()
axes2=plt.subplot(222)
plt.grid(linewidth=3)
axes3=plt.subplot(223)
#plt.grid(color='#ffbbcc')
plt.grid(color=np.random.random(3))

axes4=plt.subplot(224)
axes4.grid(axis='x',alpha=0.1,color='#ff00ff')
```
