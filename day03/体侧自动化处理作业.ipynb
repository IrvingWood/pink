### 处理不同体侧科目的分数

#### 1)处理肺活量

```python
FHL_df=score["男肺活量"]
```

```python
def boy_FHL(x):
    # 确定接收到的原始数据对应的成绩并返回

    for data in FHL_df["成绩"]:
        
        if x >= data:
            return FHL_df["分数"].loc[FHL_df["成绩"]==data].values[0]
    return 0
```

#### 3）1000米成绩的处理

```python
# 先把评分表和原始数据表的1000米成绩的数据类型进行统一
# score:3′30″
```

```python
score["男1000"].loc[1,"成绩"]
```

```python
'3\'35"'.strip('"')
```

```python
"3'35".split("\'")
```

```python
# 时间的转换方式
def process_data(x):
    # 如果不是字符串返回原始值
    if not isinstance(x,str):
        return x
    x=x.strip('"')
    m,s=x.split("\'")
    s=float(s)/60
    return float(m)+s
```

```python
process_data('3\'35"')
```

```python
grade["1000米"].loc[1]
```

```python
process_data("4'16")
```

```python
# 处理grade和score表的时间列数值
grade["1000米"]=grade["1000米"].astype(str)
```

```python
# 处理grade和score表的时间列数值
grade["1000米"]=grade1["1000米"]
```

```python
grade["1000米"]=grade["1000米"].map(process_data)
```

```python
score[("男1000","成绩")]=score[("男1000","成绩")].map(process_data)
```

```python
score.head()
```

```python
df_1000M=score["男1000"]
def map_1000M(x):
    if x==0:
        return 0
    for data in df_1000M["成绩"]:
        if x<=data:
            return df_1000M.loc[df_1000M["成绩"]==data].values[0][1]
    return 0
```
