使用std（）函数可以求得DataFrame对象每一列的标准差

```python
# 首先尊重业务的需求

# 异常值通用的界定办法：如果数据是呈标准正态分布的，|data| > 3*|data.std()|
# 离群点的检测：数值型数据都可以使用离群点的方式来检测异常

```

```python
data= np.random.randn(1000)
```

```python
data[np.abs(data) >3*data.std()]
```

```python
df = DataFrame(data=np.random.randn(1000,3), columns=list("ABC"))
df
```

```python
df.std()
```

```python
# 任意一行至少存在一个数的绝对值大于该数所处列的三倍标准差，即认定为满足异常值条件
condition = (np.abs(df) <3*df.std()).all(axis=1)
```

```python
df.loc[condition]
```

```python
# 分组对象的聚合函数，接收的参数是每一个分组
# 定制一个最大值和平均值的差的函数
def group_function(x):

    return x.max()-x.mean()
```

```python
df = DataFrame(h1.groupby("角色")["法术/\n怒气"].apply(group_function))
df.columns = ["max-mean"]
```

```python
df
```

```python
pd.merge(h1, df, left_on="角色",right_index=True)
```

```python
DataFrame(h1.groupby("角色")["法术/\n怒气"].apply(group_function))
```

```python
DataFrame(h1.groupby("角色")["法术/\n怒气"].transform(group_function))
```
